```python
import json
import os
from google.cloud import storage
from google.oauth2 import service_account

creds_path = r"C:\Users\pc\OneDrive\Escritorio\GCP- Data Engineer\gcp-data-engineer\conection-drive\credentials.json"
creds = service_account.Credentials.from_service_account_file(creds_path)

# Try to read project_id from the service account JSON; fall back to env var if missing
with open(creds_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
project_id = data.get('project_id') or os.environ.get('GOOGLE_CLOUD_PROJECT')
if not project_id:
    raise ValueError('Project id not found in credentials.json and GOOGLE_CLOUD_PROJECT is not set.')

class StorageCreated:

    def created_buckets(self):
        """Lista todos los buckets en el proyecto."""
        # Use project_id read from the credentials JSON
        client = storage.Client(project=project_id, credentials=creds)
        try:
            # The name for the new bucket
            bucket_name = "my-new-bucket-prueba"
            # Initialize a Bucket object before setting attributes
            bucket = client.bucket(bucket_name)
            bucket.storage_class = "STANDARD"

            # Creates the new bucket on the server from the Bucket object
            bucket = client.create_bucket(bucket, location="us-central1")

            print(f"Bucket {bucket.name} created.")
        except Exception as e:
            print('Error al crear buckets:', e)
            return


if __name__ == "__main__":
    created = StorageCreated()
    created.created_buckets()
```
